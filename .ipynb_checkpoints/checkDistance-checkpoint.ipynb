# Check Distance

Estimates the distance of a single stereo image, for manual verification.

```python
from urb import *
%matplotlib inline  
```

Read the left and right image.

```python
links = Frame('/data/urbinn/datasets/Na_meten/foto_4/foto_4/left/436716469974.png')
```

```python
links.read_corresponding('/data/urbinn/datasets/Na_meten/foto_4/foto_4/right')
```

```python
links.get_framepoints()
links.compute_depth()
links.filter_non_stereo()
```

```python
show(draw_framepoints_d3(links.get_framepoints()))
```

Show the estimated depth and corresponding confidence of the points.

```python
for f in links.get_framepoints():
    print(f.get_depth(), f.confidence)
```
